## Section 1: Variables and Data Types
## セクション1：変数とデータ型

Variables store information. Python has several basic data types:
- **int** (integers): whole numbers like 100, -5
- **float** (decimals): numbers with decimals like 3.14, -0.5
- **str** (strings): text in quotes like "Hello", "ABC Corp"
- **bool** (booleans): True or False

変数は情報を保存します。Pythonにはいくつかの基本的なデータ型があります：
- **int**（整数）：100、-5のような整数
- **float**（浮動小数点数）：3.14、-0.5のような小数
- **str**（文字列）："Hello"、"ABC Corp"のような引用符内のテキスト
- **bool**（ブール値）：TrueまたはFalse

### Example: Creating Variables | 例：変数の作成

```python
# Example: Company information variables
# 例：企業情報の変数

company_name = "TechVision株式会社"  # String (文字列)
employee_count = 450                   # Integer (整数)
avg_salary = 4500000.50               # Float (浮動小数点数)
is_hiring = True                      # Boolean (ブール値)

# Print the variables
print("Company:", company_name)
print("Employees:", employee_count)
print("Average Salary:", avg_salary)
print("Currently Hiring:", is_hiring)
```
